## Основные теоретические положения

### Частота Найквиста

Если гармонический сигнал имеет частоту $$f_N<\frac{f_d}{2}=\frac{1}{2T},$$ где $f_N$ - частота гармонического сигнала, $f_d$ - частота дискретизации, то тогда имея дискретный сигнал, можно восстановить аналоговый без потерь информации. В случае равенства частота сохраняется, но может измениться амплитуда сигнала и его фаза. В случае $f_N>\frac{f_d}{2}=\frac{1}{2T}$, восстановить сигнал без потерь информации не получится. $\frac{f_d}{2}$ и называют частотой Найквиста.

### Виды детерминированных дискретных сигналов

#### Единичный цифровой импульс

$$\delta_d(k)=
\begin{cases}
1, & k=0,\\
0, & k\neq0.
\end{cases}
,k=...,-2,-1,0,1,2,...$$
Используется как аналог дельта-функции:
$$\delta(t)=
\begin{cases}
\infty, & t=0,\\
0, & t\neq0.
\end{cases}
,\int_{-\infty}^{\infty}\delta(t)dt=1$$

#### Дискретный единичный скачок:
$$\sigma_d(k)=
\begin{cases}
1, & k\geq0,\\
0, & k<0.
\end{cases}
,k=...,-2,-1,0,1,2,...$$
Используется как дискретный аналог функции единичного скачка:
$$\sigma(t)=
\begin{cases}
1, & t>0,\\
0.5 & \text{or undefined,} & t=0, \\
0, & t<0.
\end{cases}$$
#### Дискретная экспоненциальная функция
$$s(k)=
\begin{cases}
0, & k<0,\\
a^k, & k\geq0.
\end{cases}
,k=...,-2,-1,0,1,2,...$$
#### Дискретная затухающая синусоида:
$$s(k)=a^kcos(k\omega+\varphi)$$
#### Дискретный комплексный гармонический сигнал:
$$s_2(k)=Cexp(j\hat\omega_0k)$$

## Выполнение работы

Пункт 1.

```python
import math
from matplotlib import pyplot as plt

# Исходные данные
Nb = 11
N = 30 + Nb % 5
T = 0.0005 * (1 + Nb % 3)
a = (-1) ** Nb * (0.8 + 0.005 * Nb)
C = 1 + Nb % 5
omega = math.pi / (6 + Nb % 5)
m = 5 + Nb % 5

# Функция единичного импульса
def unit_impulse(k):
    return 1 if k == 0 else 0

# Дискретное и дискретное нормированное время
discrete_time = [n * T for n in range(N)]
normalized_discrete_time = [n for n in range(N)]

# Графики импульсов
dt_impulse = [unit_impulse(t) for t in discrete_time]
ndt_impulse = [unit_impulse(t) for t in normalized_discrete_time]


# Функция отрисовки
def show_plot(X, Y):
    plt.clf()
    plt.grid()
    plt.plot(X, Y, 'o-')
    plt.show()
    
show_plot(discrete_time, dt_impulse)
show_plot(normalized_discrete_time, ndt_impulse)
```

a.  
$nT$ - дискретное время, где $T$ - период дискретизации  
$n$ - дискретное нормированное время, $Т = 1$  
b.  
Дельта функция в точке 0 равна бесконечности  
Единичным цифровой импульс в точке 0 равен 1

Пункт 2.

```python
# Функция единичного скачка
def unit_step(k):
    return 1 if k >= 0 else 0

# Графики скачков
dt_step = [unit_step(t) for t in discrete_time]
ndt_step = [unit_step(t) for t in normalized_discrete_time]

show_plot(discrete_time, dt_step)
show_plot(normalized_discrete_time, ndt_step)
```

а.  
Дискретная функция единичного скачка равна 1 в точке 0. Дискретная функция определена на множестве целых чисел.  
Аналоговая фукнция единичного скачка равна 0.5 или неопределена.  Аналоговая функция определена на множестве действительных чисел.  
b.  
Частота дискретизации $f_d = \frac{1}{T}$:

```python
1/T
```

$f_d = 666\frac{2}{3}$

Пункт 3.

```python
def discrete_exp(k):
    return 0 if k < 0 else (a ** k)

dt_exp = [discrete_exp(t) for t in discrete_time]
ndt_exp = [discrete_exp(t) for t in normalized_discrete_time]


show_plot(discrete_time, dt_exp)
show_plot(normalized_discrete_time, ndt_exp)
```

Дискретная экспоненциальная функция равна 0 при значениях аргумента меньших 0.  
Аналоговая экспоненциальная функция не равна 0 при значениях аргумента меньших 0, и равна $a^x$ при всех значениях аргумента.

Пункт 4.

```python
import numpy

def discrete_harmonic(k):
    return C * numpy.exp(1j * omega * k)

ndt_dt_harmonicexp = [discrete_harmonic(t) for t in normalized_discrete_time]

show_plot(normalized_discrete_time, numpy.real(ndt_dt_harmonicexp))
show_plot(normalized_discrete_time, numpy.imag(ndt_dt_harmonicexp))
```

$C \cdot e^{j \hat{\omega}_0 k} = C \cdot cos(\hat{\omega}_0 k) + j \cdot C \cdot sin(\hat{\omega}_0 k)$ - Формула Эйлера  
$C \cdot cos(\hat{\omega}_0 k)$ - действительная часть  
$C \cdot sin(\hat{\omega}_0 k)$ - мнимая часть

Пункт 5.

```python
# Сдвиг последовательности на m
delayed_discrete_time = [(n - m) for n in range(N)]
# Задержанные графики
delayed_impulse = [unit_impulse(t) for t in delayed_discrete_time]
delaed_step = [unit_step(t) for t in delayed_discrete_time]
delayed_exp = [discrete_exp(t) for t in delayed_discrete_time]

show_plot(discrete_time, delayed_impulse)
show_plot(discrete_time, delaed_step)
show_plot(discrete_time, delayed_exp)
```

$$
